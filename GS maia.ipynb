```python
import openai
from google.colab import userdata

client = openai.OpenAI(api_key=userdata.get('OPENAI_API_KEY'))

def chatbot_aca(usuario, mensagem_usuario):
    # Definir o papel do assistente
    if usuario == 1:
        sistema = (
            "Solicite que o usuário descreva qual é o seu trabalho atual e liste as principais atividades que realiza no dia a dia. A partir dessas informações, será possível avaliar o nível de exposição da função ao risco de automação e, com isso, apontar quais competências profissionais tendem a se tornar mais importantes para garantir relevância e segurança no futuro dessa área."
        )
    elif usuario == 2:
        sistema = (
            "Considerando a profissão informada e as habilidades essenciais relacionadas a ela, apresente de três a cinco áreas em que o usuário pode concentrar seus esforços de desenvolvimento profissional. Para cada uma dessas áreas de aprimoramento, indique também um recurso de aprendizado — como cursos, materiais de estudo ou ferramentas práticas — que possa auxiliar o usuário a evoluir de forma estruturada e eficaz. "
        )
    elif usuario == 3:
        sistema = (
            "Elabore uma recomendação completa de reskilling para o usuário, indicando caminhos que possam facilitar sua transição para novas oportunidades profissionais. Para isso, identifique primeiro quais habilidades transferíveis ele já possui — aquelas competências que podem ser aproveitadas em diferentes funções e setores. Em seguida, apresente de três a cinco novas habilidades que ele deveria desenvolver para ampliar suas chances de recolocação no mercado de trabalho, considerando tendências atuais e demandas emergentes"
        )
    elif usuario == 4:
        sistema = (
            "Conduza uma breve entrevista com o usuário composta por três perguntas essenciais, elaboradas para avaliar sua adequação a uma nova função. As perguntas devem priorizar clareza, relevância para o cargo em questão e profundidade suficiente para revelar sua experiência, capacidades e expectativas profissionais."
        )

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": sistema},
            {"role": "user", "content": mensagem_usuario}
        ],
        temperature=0.3,
        max_tokens=600
    )

    # Acesso do conteúdo do objeto de resposta de v1
    resposta = response.choices[0].message.content
    return resposta

print("Agente IA de Carreira Adaptativa")
print("1. Análise de Perfil e Risco de Automação.")
print("2. Plano de Upskilling Personalizado.")
print("3. Sugestão de Caminho de Reskilling")
print("4. Simulação de Entrevista \n")

entry = int(input("Escolha uma opção: "))

while True:
    user_input = input("\nVocê: ")

    if user_input.lower() in ["sair"]:
        print("Encerrando o agente IA.")
        break

    resposta = chatbot_aca(entry, user_input)
    print(f"\n Agente: {resposta}")
```
